# Modelo

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.constraints import max_norm

print(tf.__version__)
```

```python
# Particiona os dados
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x_data, y_val, test_size=0.1, shuffle= True)
```

```python
# %load_ext tensorboard

# import datetime, os
# logs_base_dir = "./logs"
# os.makedirs(logs_base_dir, exist_ok=True)
# %tensorboard --logdir {logs_base_dir}
```

```python
# https://stats.stackexchange.com/questions/181/how-to-choose-the-number-of-hidden-layers-and-nodes-in-a-feedforward-neural-netw
# https://www.tensorflow.org/tutorials/keras/regression?hl=pt-br

def constroiModelo(n):
  modelo = keras.Sequential([
    layers.Dense(round(4*n/7) + 1, activation='relu', input_shape=[len(x_data.keys())]),
    layers.Dropout(0.2),
    layers.Dense(round(4*n/14) + 1, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(round(4*n/28) + 1, activation='relu'),
    layers.Dense(1)
  ])

  otimizador = tf.keras.optimizers.Nadam(0.001)
  # otimizador = tf.keras.optimizers.RMSprop(0.001)
  # otimizador = tf.keras.optimizers.Adamax(0.001)
   
  # https://keras.io/api/optimizers/adam/
  modelo.compile(loss='mae', # FunÃ§Ã£o a minimizar
                 optimizer=otimizador,
                 metrics=['mse', 'accuracy'])

  return modelo
```

```python
import numpy as np
import time
from sklearn.model_selection import train_test_split

def processaModelo(x, y, n):
    inicio = time.time()
    modelo = constroiModelo(n)
    
    # Particiona os dados
    x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.1, shuffle= True)
    
    # modelo.summary()
    processamento = modelo.fit(x = x_treino, 
                             y = y_treino, 
                             epochs = 1500,
                             # validation_split = 0.1,
                             validation_data=(x_teste, y_teste), 
                             batch_size = 400,
                             verbose = 0
                             )
    print("Processamendo n: ", n, ', com ', time.time()-inicio)
    return processamento
    #return modelo
```

```python
resultados = pd.DataFrame(columns=['n', 'mae'])
```

```python
inicio = time.time()
for aux in range(12):
    df = pd.DataFrame(range(5, 300, 5), columns=['n'])
    df['mae'] = df.apply(lambda row : np.min(processaModelo(x_data, y_val, row[0]).history['val_loss']), axis = 1)

    print(aux + 1, " - Processamendo total: ", time.time()-inicio)
    resultados = resultados.append(df)
```
